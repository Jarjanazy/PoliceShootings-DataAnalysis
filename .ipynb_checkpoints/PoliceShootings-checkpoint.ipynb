```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('PoliceShootings.csv')
```

```python
df['Month']  = df.apply(lambda row : row['date'].split('/')[1], axis=1)
df['Year'] = df.apply(lambda row : row['date'].split('/')[2], axis=1)
df.drop('date', axis=1, inplace=True)
```

```python
df.head()
```

```python
plt.figure(figsize=(8, 4))
ax = sns.countplot(x="race", hue="gender",data=df, order=["White", "Black", "Hispanic", "Asian", "Native", "Other"])
ax.set(ylabel="Count of victims", xlabel="Race of victim", title="Count of victimes per race")
plt.tight_layout()
```

```python
plt.figure(figsize=(20,15))
ax = sns.countplot(y="state", data=df, orient="h", order= df['state'].value_counts().index)

ax.set_yticklabels(ax.get_yticklabels(), fontsize=20)

plt.tight_layout()
```

```python
plt.figure(figsize=(8, 4))
sns.violinplot(x="race", y="age", hue="gender", data=df)
plt.tight_layout()
```

```python
ax = sns.distplot(df['age'], kde=False)
ax.set(title="Age distribution of victims", ylabel="Count of victims")
plt.tight_layout()
```

```python
order = df['armed'].value_counts()[:5].index
ax= sns.countplot(x='armed',data=df, orient="h",  order= order)
ax.set(title="Weapons carried by victims")
plt.tight_layout()
```
